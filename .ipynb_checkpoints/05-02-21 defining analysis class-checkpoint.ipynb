```python
from UniDec import unidec
from UniDec.unidec_modules import unidectools as ud
from UniDec.metaunidec import mudeng 
from UniDec.metaunidec import meta_import_wizard as wiz
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.gridspec as gridspec
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
from ctypes import *
import sys
import os
import baf2sql
from pathlib import Path
import time 
```

```python
def find_nearest(array, value):
    """Returns index of float/int closest to value you want to find"""
    array = np.asarray(array)
    return (np.abs(array - value)).argmin()
```

```python
class MassSpecAnalysis(mudeng.MetaUniDec):
    def __init__(self):
        
        mudeng.MetaUniDec.__init__(self)
    
    def get_mean_spectra(self, rt_range, baf_fn, raw_calibration):
        """Means spectra within a time window
        
        Returns mean profile mz and int"""
        
        
        spec_vals = {'profile_mz':[], 'profile_int':[]}
        sqlite_fn = baf2sql.getSQLiteCacheFilename(baf_fn)
        conn = sqlite3.connect(sqlite_fn)
        
        for i in rt_range: 
            q = conn.execute("SELECT LineMzId, LineIntensityId, ProfileMzId, ProfileIntensityId FROM Spectra "
                            "WHERE ABS(Rt - {}) < 1e-8".format(i))
                        
            row = q.fetchone()

            bs = baf2sql.BinaryStorage(baf_fn)

            if not all(row) == False: # check for None values

                bs = baf2sql.BinaryStorage(baf_fn)

                profile_mz = bs.readArrayDouble(row[2])
                profile_int = bs.readArrayDouble(row[3])

                spec_vals['profile_mz'].append(profile_mz)
                spec_vals['profile_int'].append(profile_int)
                
            else:
                print("No spectra found at {:.2f} seconds".format(i))
                
        mean_profile_mz = np.average(np.array(spec_vals['profile_mz']), axis = 0)
        mean_profile_int = np.average(np.array(spec_vals['profile_int']), axis = 0)
        
        return [mean_profile_mz, mean_profile_int]
    
    def upload_bruker(self, analysis_dir, first_peak = None, spacing = None, time_window = None, num_peaks = None,
                      plot_LC = False, raw_calibration = False):
        
        self.analysis_dir = analysis_dir
        self.first_peak = first_peak
        self.spacing = spacing
        self.time_window = time_window
        self.num_peaks = num_peaks
        
        tstart = time.perf_counter()
        
        # check time window
        if time_window != None and time_window % 2 != 0: 
            raise ValueError("time_window must be even or None")
            
        # --- data extraction using baf2sql DLL 

        # if sys.version_info.major == 2:
        #     # note: assuming a european Windows here...
        #     analysis_dir = unicode(analysis_dir, 'cp1252') #???

        baf_fn = os.path.join(analysis_dir, "analysis.baf")
        sqlite_fn = baf2sql.getSQLiteCacheFilename(baf_fn)
        conn = sqlite3.connect(sqlite_fn)

        # --- Count spectra
        q = conn.execute("SELECT COUNT(*) FROM Spectra "
                        "WHERE LineMzId NOT NULL AND ProfileMzId NOT NULL")
        row = q.fetchone()
        N = row[0]
        print("Specified BAF has {} spectra with line and profile data.".format(N))

        # --- Plot TIC and BPC over MS^1 spectra
        q = conn.execute("SELECT Rt, SumIntensity, MaxIntensity FROM Spectra s "
                        "JOIN AcquisitionKeys ak ON s.AcquisitionKey = ak.Id "
                        "WHERE ak.MsLevel = 0 "
                        "ORDER BY s.ROWID")
        data = [ row for row in q ]
        rt = [ row[0] for row in data ]
        tic = [ row[1] for row in data ]
        bpc = [ row[2] for row in data ]
        
        self.rt, self.tic, self.bpc = rt, tic, bpc
        # plot LC
        if plot_LC == True:

            plt.figure()
            plt.plot(rt, tic)
            plt.plot(rt, bpc, '--')
            plt.xlabel('retention time / s')
            plt.ylabel('intensity / a.u.')
            plt.legend(['TIC', 'BPC'])
            plt.show()
        
        
        np_bpc = np.array(bpc, dtype = 'float')
        
        # --- Plot mean spectrum within time window of highest peak in LC
        if self.first_peak == None:          

            # get time window around base-peak intensity 
            rt_range =  rt[np.nanargmax(np_bpc) - int(time_window/2): np.nanargmax(np_bpc) + int(time_window/2 + 1)]

            self.raw_data = self.get_mean_spectra(rt_range, baf_fn, raw_calibration)
            
            
        # --- Plot mean spectrums within time windows at stipulated peaks 
        elif self.first_peak != None and self.spacing != None and self.num_peaks != None: 
            
            self.peak_times = [self.first_peak + i*self.spacing for i in range(self.num_peaks)]
            
            # set up dictionary to house data
            self.raw_data = {}
            
            # loops through peaks, calculates mean spectrum within time window for each and store self.raw_data[x]
            # where x is time in s of the peak 
            
            for i in range(self.num_peaks):
            
                # get nearest time point
                nearest = find_nearest(rt, self.peak_times[i])

                # get range for which to mean over
                rt_range = rt[(nearest - int(self.time_window/2)):(nearest + int(self.time_window/2 + 1))]
                
                self.raw_data[self.peak_times[i]] = self.get_mean_spectra(rt_range, baf_fn, raw_calibration)
        
        # --- Splice entire data into time windows and take mean of each  (this might end up in a different function)
        elif self.spacing != None and self.first_peak == None and self.num_peaks == None :
            pass
            
            
```

```python
# MetaUniDec
eng = MassSpecAnalysis()
```

```python
eng.upload_bruker("3x_injection.d", time_window = 10, plot_LC = True, first_peak = 0, spacing = 10, num_peaks = 30)
```

```python
for rawdata in eng.raw_data.items():
    plt.figure()
    plt.plot(rawdata[1][0], rawdata[1][1])
    plt.title("Peak at {}".format(rawdata[0]))
plt.show()
```
